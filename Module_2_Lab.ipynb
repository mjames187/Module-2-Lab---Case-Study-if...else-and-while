```python
# Madeline James
# Dean's List and Honor Roll Checker
# Allows the user to input a student's name and GPA, then determines whether or not that student qualifies for the Dean's List and/or the Honor Roll.
gather_records = True
while gather_records:
    last_name = input("What is the student's last name? Enter \"ZZZ\" to cancel.")
    if last_name == "ZZZ":
        gather_records = False
        break
    first_name = input("What is the student's first name?")
    gpa = float(input("What is the student's GPA?"))
    if gpa >= 3.5:
        print(first_name, last_name, "has made the Dean's List.")
    if gpa >= 3.25:
        print(first_name, last_name, "has made the Honor Roll.")
```
